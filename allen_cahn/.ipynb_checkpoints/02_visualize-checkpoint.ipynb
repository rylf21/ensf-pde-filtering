# Allen--Cahn: Visualizations

Loads NPZ results from `01_run_experiment.ipynb` and produces
the thesis figures (Chapter 3).

**Figures produced:**
- Field snapshots at selected times (truth vs EnSF estimate)
- RMSE over time for 100% and 70% observation densities
- Supremum norm comparison (estimated vs reference)
- Discrete energy over time

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

DATA_DIR    = Path("../data")
FIGURES_DIR = Path("../figures")
FIGURES_DIR.mkdir(parents=True, exist_ok=True)

FILE_FORMAT = "pdf"
```

## Load results

```python
def load(tag):
    return dict(np.load(DATA_DIR / f"{tag}.npz", allow_pickle=True))

truth   = load("ac_dataset_case1")
res100  = load("ac_ensf_obs100")
res70   = load("ac_ensf_obs70")

X_true  = truth["X_true_obs"]   # (Nobs, state_dim)
t_obs   = truth["t_obs"]
Nx      = int(truth["Nx"])
Ny      = int(truth["Ny"])

state_dim = X_true.shape[1]
print(f"State dim: {state_dim}, Nobs: {len(t_obs)}")
```

## Coordinate mapping (DOF → 2D image)

For Q1 on a structured quad mesh the DOF ordering matches a regular grid,
so we can reshape for imshow-style plots.

```python
# Q1 on (Nx x Ny) quad mesh → (Nx+1) x (Ny+1) nodes
nx_img = Nx + 1
ny_img = Ny + 1
assert nx_img * ny_img == state_dim

def to_img(vec):
    """Reshape DOF vector to 2D image (ny, nx)."""
    return vec.reshape(ny_img, nx_img)

extent = [-0.5, 0.5, -0.5, 0.5]
```

## Figure 1: Field snapshots

```python
def plot_snapshot_row(res, step_indices, tag):
    """
    Plot truth and EnSF estimate at selected time steps.
    step_indices: list of 1-indexed assimilation steps.
    """
    n = len(step_indices)
    fig, axes = plt.subplots(2, n, figsize=(3.5 * n, 7))

    vmin, vmax = -1.0, 1.0
    cmap = "RdBu_r"

    for col, k in enumerate(step_indices):
        t_k = t_obs[k]

        axes[0, col].imshow(
            to_img(X_true[k]), origin="lower", extent=extent,
            cmap=cmap, vmin=vmin, vmax=vmax
        )
        axes[0, col].set_title(f"Truth, $t={t_k:.1f}$", fontsize=10)
        axes[0, col].axis("off")

        im = axes[1, col].imshow(
            to_img(res["EnSF_est"][k]), origin="lower", extent=extent,
            cmap=cmap, vmin=vmin, vmax=vmax
        )
        axes[1, col].set_title(f"EnSF, $t={t_k:.1f}$", fontsize=10)
        axes[1, col].axis("off")

    fig.colorbar(im, ax=axes, orientation="vertical", fraction=0.02, pad=0.04)
    fig.tight_layout()

    out = FIGURES_DIR / f"ac_{tag}_snapshots.{FILE_FORMAT}"
    fig.savefig(out, bbox_inches="tight", format=FILE_FORMAT)
    print(f"Saved: {out}")
    plt.show()

# Snapshots at t=0, 2.5, 7.5, 10  (steps 0, 62, 187, 249)
steps = [0, 62, 187, 249]
plot_snapshot_row(res100, steps, tag="obs100")
plot_snapshot_row(res70,  steps, tag="obs70")
```

## Figure 2: RMSE, supremum norm, energy

```python
fig, axes = plt.subplots(1, 3, figsize=(13, 4))

t_plot = t_obs[1:]

# RMSE
axes[0].plot(t_plot, res100["step_rmse"], label="100% obs", color="steelblue")
axes[0].plot(t_plot, res70["step_rmse"],  label="70% obs",  color="darkorange")
axes[0].set_xlabel("Time $t$", fontsize=12)
axes[0].set_ylabel("RMSE", fontsize=12)
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# Supremum norm
axes[1].plot(t_plot, res100["supnorm_arr"], label="EnSF 100%", color="steelblue")
axes[1].plot(t_plot, res70["supnorm_arr"],  label="EnSF 70%",  color="darkorange")
axes[1].plot(t_obs[1:], truth["supnorm"][1:],  label="Reference",  color="k", linestyle="--")
axes[1].set_xlabel("Time $t$", fontsize=12)
axes[1].set_ylabel(r"$\|\phi\|_\infty$", fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

# Energy
axes[2].plot(t_plot, res100["energy_arr"], label="EnSF 100%", color="steelblue")
axes[2].plot(t_plot, res70["energy_arr"],  label="EnSF 70%",  color="darkorange")
axes[2].plot(t_obs[1:], truth["energies"][1:], label="Reference", color="k", linestyle="--")
axes[2].set_xlabel("Time $t$", fontsize=12)
axes[2].set_ylabel("Energy $E(\phi)$", fontsize=12)
axes[2].legend(fontsize=10)
axes[2].grid(True, alpha=0.3)

fig.tight_layout()
out = FIGURES_DIR / f"ac_diagnostics.{FILE_FORMAT}"
fig.savefig(out, bbox_inches="tight", format=FILE_FORMAT)
print(f"Saved: {out}")
plt.show()
```
